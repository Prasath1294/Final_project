# Outliers Detection

```python
#to find Outliers
lesser=[]
greater=[]

for columnName in quan:
    if (descriptive[columnName]["Lesser"]>descriptive[columnName]["Min"]):
        lesser.append(columnName)
    if (descriptive[columnName]["Greater"]<descriptive[columnName]["Max"]):
        greater.append(columnName)
```

```python
lesser
```

# Replacing Outliers

```python
#replacing outliers with lesser and greater values
for column in lesser:
    dataset[column][dataset[column]<descriptive[column]["Lesser"]]=descriptive[column]["Lesser"]
for column in greater:
    dataset[column][dataset[column]>descriptive[column]["Greater"]]=descriptive[column]["Greater"]
```

```python
independent = dataset[['id', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

dependent = dataset['diagnosis_M']
```

# Feature Selection

```python
def selectkbest(independent,dependent,n):
        test = SelectKBest(score_func=chi2, k=n)
        fit1= test.fit(independent,dependent)
        selectk_features = fit1.transform(independent)
        return selectk_features
```

```python
kbest=selectkbest(independent,dependent,5)
```

```python
kbest
```

# Training , Testing and Splitting Data

```python
X_train,X_test,y_train,y_test = train_test_split(kbest, dependent, test_size= 1/3, random_state=0)
```

# Standardisation

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Model Creation on Support Vector Classifier

```python
param_grid = {'kernel':['linear','rbf','poly','sigmoid'],'gamma':['auto','scale'],'C':[1,10,100,1000]}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3, n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train, y_train)
```

```python
grid_prediction = grid.predict(X_test)
```

```python
re = grid.cv_results_
```

```python
cm = confusion_matrix(y_test, grid_prediction)
```

```python
cls_reports = classification_report(y_test, grid_prediction)
```
